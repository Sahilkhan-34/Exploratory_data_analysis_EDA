```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv(r'Earthquakes in pakistan.csv', encoding='latin1')
df.head()
```

```python
df.rename(columns={'Locality, district, or province':'Locality'},inplace=True)
df
```

```python
df.info()
```

```python
for i in df['Date']:
    df['Date']= i.replace('\n','')
df['Date']
```

```python
df['Date'] = df['Date'].fillna('2000-01-01')
```

```python
df['Date']
```

```python
from datetime import datetime
for i in df['Date']:
    # Your date string
    date_string = i

    # Remove the newline character
    date_string = date_string.replace('\n', '')

    # Convert to a datetime object
    date_object = datetime.strptime(date_string, '%Y-%m-%d')

    # Print the date in a normal format
    print(date_object.strftime('%Y-%m-%d'))
df['Date']
```
